```python
import pandas as pd
import requests
from tabulate import tabulate
import os
import csv
from pandas import *

import yagmail
from datetime import date

from google.colab import files as FILE
from PIL import Image

from bs4 import BeautifulSoup
import numpy as np
from time import sleep
from random import randint
```

```python
url = 'https://knowyourmeme.com/memes/popular/page/' +str(randint(1, 95))

headers = {'User-Agent': 'Mozilla/5.0 (X11; Linux x86_64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/79.0.3945.117 Safari/537.36'}

r = requests.get(url, headers=headers)
soup = BeautifulSoup(r.content, 'html.parser')

url
```

```python


memes = soup.find_all('a', {'class': 'photo'})

ls = []

for meme in memes:
  dictm={}
  desc = meme.find('img')
  if desc:
    dictm['Title'] = desc['title']
  urlm = meme.find('img', {"data-src":True})
  if urlm:
    dictm['URL'] = urlm['data-src']
  ls.append(dictm)




df1 = pd.DataFrame(ls, columns=['URL', 'Title'])
df1 = df1.dropna(subset = ['URL'])
df2 = df1[df1['URL'].str.contains("entries")]

df2
```

```python
#medium = df2[df2['URL'].str.replace('medium','mobile')]

df2["URL"] = df2["URL"].apply(lambda x: x.replace('medium','mobile'))
df2["URL"] = df2["URL"].apply(lambda x: x.replace('png','jpg'))

final = pd.DataFrame(df2, columns=['URL', 'Title'])
final
```

```python
#formating

meme1 = final.sample()
# meme2 = final.sample()
meme_url = str(meme1['URL'].to_csv(index=False, header=None))
meme_url = meme_url[:-1]

meme_name = str(meme1['Title'].to_csv(index=False, header=None))
meme_name = meme_name[:-1]

# path2 = memeurl2[:-1]
print(meme_name, meme_url)
```
